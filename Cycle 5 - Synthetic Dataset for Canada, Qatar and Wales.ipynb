```python
import pandas as pd
```

# 1. Load Data

```python
df1 = pd.read_csv('./datasets/international_matches_raw.csv')
```

```python
df1.tournament.unique()
```

```python
new_countries = ['Canada', 'Qatar', 'Wales']
df11 = pd.DataFrame()
df11 = pd.concat([df11, df1[df1.home_team.isin(new_countries)], df1[df1.away_team.isin(new_countries)]])
```

```python
df11 = df11[['date', 'home_team', 'away_team', 'home_team_score', 'away_team_score']]
```

```python
df11.columns = ['Datetime', 'Home Team Name','Away Team Name', 'Home Team Goals', 'Away Team Goals']
```

```python
df11['Datetime'] = pd.to_datetime(df11['Datetime'])
df11['Year'] = df11['Datetime'].dt.year
```

```python
cols_order = ['Year', 'Datetime', 'Home Team Name','Away Team Name', 'Home Team Goals', 'Away Team Goals']
df11 = df11[cols_order].reset_index(drop=True)
```

```python
df_final = pd.DataFrame()
```

```python
df_final = pd.concat([df_final, df11[df11.Year == 2014], df11[df11.Year == 2018]])
```

```python
df_final.shape
```

```python
df_final['Year'] = df_final.Year.astype('Int64')
```
