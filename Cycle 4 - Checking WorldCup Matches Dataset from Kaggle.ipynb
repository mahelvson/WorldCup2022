# 0. Imports

```python
import pandas as pd
```

# 1. Data Loading

```python
df_raw = pd.read_csv('./datasets/WorldCupMatches_raw.csv')
```

```python
df_raw.shape
```

```python
columns_filter = ['Year',
                  'Datetime',
                  #'Stage',
                  #'Stadium',
                  #'City',
                  'Home Team Name',
                  'Away Team Name',
                  #'Win conditions',
                  #'Attendance',
                  #'Half-time Home Goals',
                  #'Half-time Away Goals',
                  #'Referee',
                  #'Assistant 1',
                  #'Assistant 2',
                  #'RoundID',
                  #'MatchID',
                  #'Home Team Initials',
                  #'Away Team Initials',
                  'Home Team Goals',
                  'Away Team Goals'
                  ]
df1 = df_raw[columns_filter]
```

```python
df1.head()
```

## 1.1 Loading 2018 Dataset

```python
df_2018 = pd.read_csv('./datasets/2018_worldcup_v3.csv')
df_2018 = df_2018[columns_filter]
df_2018.head()
```

```python
df1_complete = pd.concat([df1, df_2018], axis = 0, ignore_index=True)
```

```python
df1_complete.head()
```

## 1.2 Loading 2022 Template

```python
df_2022 = pd.read_csv('./datasets/2022_world_cup_template.csv')
```

```python
df_2022['Datetime'] = pd.to_datetime(df_2022.Datetime, dayfirst=False, infer_datetime_format=True)

def format_time(row):
    return row['Datetime'].strftime('%d %b %Y - %H:%M')

df_2022['Datetime'] = df_2022.apply(format_time, axis=1)
```

```python
df_2022
```

```python
df1_complete2 = pd.concat([df1_complete, df_2022], axis = 0, ignore_index=True)
```

# 2. Creating Dataset

```python
df2 = df1_complete2.copy()
```

```python
df = df2.drop_duplicates()
```

```python
df2.to_csv('./datasets/WorldCupMatches.csv')
```
