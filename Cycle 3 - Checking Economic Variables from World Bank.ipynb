# 2. Data Manipulation

```python
df2 = df_economic_raw.drop('Series Code', axis=1)
```

```python
df2.columns = ['Country Name', 'Country Code', 'Series Name', 'YR1993',
               'YR1997', 'YR2001', 'YR2005', 'YR2009',
               'YR2013', 'YR2017', 'YR2020']
```

```python
def clean_series_name(row):
    if row['Series Name'] == 'GDP (current US$)':
        row['Series Name'] = 'GDP'
    elif row['Series Name'] == 'Population, total':
        row['Series Name'] = 'POP'
    elif row['Series Name'] == 'Adjusted net national income per capita (current US$)':
        row['Series Name'] = 'INC'
    
    return row['Series Name'] 

df2['Series Name'] = df2.apply(clean_series_name, axis=1)
```
